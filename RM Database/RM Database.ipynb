```python
import pandas as pd
import numpy as np

df_demo=pd.read_csv('RM Demo.csv')
df_demo
```

```python
df_alcohol=pd.read_csv('RM_EtOH.csv')
print(df_alcohol.columns)
df_alcohol=df_alcohol[['mrn','drinks_per_week']]

df=df_demo.merge(df_alcohol,on='mrn',how='outer')
print (df.dropna(subset=['drinks_per_week']))
```

```python
df_tobacco=pd.read_csv('RM_tobacco.csv')
print(df_tobacco.columns)
df_tobacco=df_tobacco[['mrn','TOBACCO_USER','pack_years']]

df=df.merge(df_tobacco,on='mrn',how='outer')
print(df.dropna(subset=['TOBACCO_USER']))
```

```python
df_stage_and_TNM=pd.read_csv('RM Staging.csv')
print(df_stage_and_TNM.columns)

df=df.merge(df_stage_and_TNM,on='mrn',how='outer')

print(df.dropna(subset=['Stage']))
```

```python
df_nar=pd.read_csv('RM_narrative.csv',error_bad_lines=False)
print (df_nar.columns)
df=df.merge(df_nar,on='mrn',how='outer')

print(df.dropna(subset=['Dx_']))
```
