```python
import pandas as pd
import numpy as np

### get the surgery info- this trashes the scaling so remove if you want a prettier pic
df_surg=pd.read_csv('RM_surgeries.csv')
#print (df_surg)
cols=['mrn','order_start_dttm']
df_surg=df_surg[cols]
df_surg.columns=['mrn','enc_date']
df_surg['is_RM']=0
print (df_surg)
```

```python
df_rad=pd.read_csv('RM_radiation_cycles.csv')
df_rad['start']=pd.to_datetime(df_rad['start'])
df_rad['end']=pd.to_datetime(df_rad['end'])
#print (df_rad)
diffs=[]
for index,row in df_rad.iterrows():
    diff=row['end']-row['start']
    diffs.append(diff.days)
df_rad['diff']=diffs

df_rad=df_rad[df_rad['diff']>30]
print (df_rad['diff'].describe())
#print(df_rad)
df_rad=df_rad[['mrn','end']]
df_rad.columns=['mrn','enc_date']
df_rad['is_RM']=0
print (df_rad)
```

```python
file='RM_MCs.csv'
df=pd.read_csv(file)
df.columns=['enc_date','mrn','is_RM']
df=df[['mrn','enc_date','is_RM']]
#print (df)

df=df[df['is_RM']==1]
print(df)

df=df.append(df_surg)
df=df.append(df_rad)
df['enc_date']=pd.to_datetime(df['enc_date'])
df=df.sort_values(by='enc_date')
print (df.shape[0])
```

```python
mrns=df.mrn.unique()
ptx_info=[]
for mrn in mrns:
    sub_df=df[df['mrn']==mrn]
    ptx_info.append(sub_df)
    
good_mrns=[]
for ptx in ptx_info:
    if 0 in ptx['is_RM'].unique() and 1 in ptx['is_RM'].unique():
        good_mrns.append(ptx.iloc[0]['mrn'])

print (good_mrns)

df=df[df['mrn'].isin(good_mrns)]
mrns=df.mrn.unique()
ptx_info=[]
for mrn in mrns:
    sub_df=df[df['mrn']==mrn]
    ptx_info.append(sub_df)
for ptx in ptx_info:
    print (ptx)
```

```python
def trim_df(df):
    df=df.reset_index(drop=True)
    if df.iloc[0]['is_RM']==1:
        keep_going=True
        i=0
        for index,row in df.iterrows():
            if row['is_RM']==0 and keep_going:
                i=index
                keep_going=False
        df_to_return=df.iloc[i:]
        return df_to_return
    else:
        return df
for index,ptx in enumerate(ptx_info):
    ptx_info[index]=trim_df(ptx)
    ptx_info[index]=ptx_info[index].drop_duplicates(keep='first',subset='enc_date')
    #print (ptx_info[index])

def fillforward(df):
    if df.iloc[0]['is_RM']==0:
        states=[]
        for index,row in df.iterrows():
            keep_going=True
            if row['is_RM']==0 and keep_going:
                states.append(0)
            elif row['is_RM']==1 and keep_going:
                states.append(1)
                keep_going=False
            else:
                states.append(1)
        df['is_RM']=states
        return df
    else:
        return df
df_rebuilt=''
has_started=False
for index,ptx in enumerate(ptx_info):
    ptx_info[index]=fillforward(ptx)
    
    try:
        first_entry=ptx_info[index].iloc[0]['enc_date']
        second_entry=ptx_info[index].iloc[1]['enc_date']
        diff=(second_entry-first_entry).days
    except:
        diff=0
    
    flag=True
    if diff<14:
        flag=False
    if flag:
        print (ptx_info[index])
    if not has_started and flag==True:
        df_rebuilt=ptx_info[index]
        has_started=True
    elif flag==True:
        df_rebuilt=df_rebuilt.append(ptx_info[index])
df_rebuilt=df_rebuilt[['enc_date','mrn','is_RM']]
df_rebuilt.columns=['date_of_service_dttm','mrn','is_RM']
df_rebuilt=df_rebuilt.set_index('date_of_service_dttm')
df_rebuilt.to_csv("RM_MCs_rad.csv")
print (df_rebuilt.shape[0])
```
