```python
import pandas as pd
import numpy as np
import datetime as dt
from dateutil.relativedelta import relativedelta

file_name='ENT_CA_DM_Cohort.txt'
df=pd.read_csv(file_name,error_bad_lines=False,delimiter='|',parse_dates=True)
print (df.columns)

### only get info for mrns associated with dataprocessing
# mrn_csv=pd.read_csv('all_mrns_of_interest.csv', parse_dates=True)
# print ('number mrns in csv:',mrn_csv.shape[0])
# mrns=mrn_csv.MRN.unique()
# print ('number mrns after formatting',len(mrns))
# df=df[df['mrn'].isin(mrns)]
# print ('number mrns in demo after filtering',df.shape[0])

### get age
df['date_of_birth']=pd.to_datetime(df['date_of_birth'], infer_datetime_format=True) 
df['date_of_death']=pd.to_datetime(df['date_of_death'], infer_datetime_format=True) 

df.to_csv('RM Demo.csv',index=False)


ages=[]
for index,row in df.iterrows():
    start_date=row['date_of_birth']
    end_date=row['date_of_death']
    try:
        diff=relativedelta(end_date, start_date).years
        ages.append(diff)
    except:
        ages.append(np.nan)
df['age']=ages
print (df.age.describe())
```
